1A.
Normal distribution is when the data tends around a central value and is symmetrical about the mean. The "bell curve" is a normal distribution. The standard normal table (Z table) is a table of all the probabilities observed in a normal distribution.

1B.
Use stats.norm.cdf() to test these three values of sigma: 1, 1.7, 2.6
The z-values should be: σ(1) = 0.84134, 
                        σ(1.7) = 0.95543
                        σ(2.6) = 0.99534

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
import seaborn as sns

#this sets the size of the plot to something useful
plt.rcParams["figure.figsize"] = (15,10)


#1B sigma = 1, 1.7, 2.6
p1 = stats.norm.cdf(1)
p2 = stats.norm.cdf(1.7)
p3 = stats.norm.cdf(2.6)

print(p1)
print(p2)
print(p3)
```

1C. 
Using the z-values from 1B, test if stats.norm.ppf() gives the same sigma values (1, 1.7, 2.6).

```python
s1 = stats.norm.ppf(p1)
s2 = stats.norm.ppf(p2)
s3 = stats.norm.ppf(p3)

print(s1)
print(s2)
print(s3)
```

2A.
Chi-square distribution is the distribution of a sum of the squares of k number of degrees of freedom. This distribution has the one parameter k, which is the degrees of freedom, and it is often used in line-fitting. 

If X_n are standard normal random variables, then Σ(X_n)^2 has a Chi-squared distribution.

Chi-square distribution is a special case of Gamma distribution.

2B.
Make a chi-square distribution plot with k (degrees of freedom) equal to 3.

```python
d = stats.chi2.rvs(loc = 5., scale = 0.01, size = 100000, df = 3)

#create plot and tick marks
fig, ax = plt.subplots(1, 1)
ax.hist(d,50, density=True)
plt.tick_params(labelsize = 20)

# make y-axis log scale
plt.yscale('log')

#set limits of x-axis
plt.xlim([5,5.2])

# label axes
plt.xlabel('x', fontsize = 20)
plt.ylabel('Probability (logscale)', fontsize = 20)

#plot graph and density function
x = np.linspace(4.95,5.2,1000)
ax.plot(x,stats.chi2.pdf(x,loc = 5., scale = 0.01, df = 3),linewidth = 8,alpha = 0.7)
plt.show()
```

1A. 
Binomial distribution is a discrete probability distribution, where there are n number of trials and p is the probability of success. The distribution is what the probability of success (p) is of having x events over n trials.

1B.
Make a series of plots with different parameters. The parameters are n (number of trials), p (probability of success of each trial) and k (number of successes). N and k will be any value equal or greater than zero and p will be between [0,1].

```python
n = 20
p = 0.7

#create plot
d = stats.binom.rvs(n, p, size = 10000)
ax = sns.displot(d, kde = False, linewidth = 20, alpha = 0.7)
plt.tick_params(labelsize = 20)

# make y-axis log scale
plt.yscale('log')

# axes and plot
plt.xlabel('x', fontsize = 20)
plt.ylabel('Probability (logscale)', fontsize = 20)
plt.show()
```

A discrete probability and sigma value means that for each independent experiment there will be a discrete probability and sigma value. The probability and sigma of one experiment is not related to that of another experiment. This is because the integral of the probability density function takes the area under the curve in the shape of rectangles, which will be a discrete value.

The mean does not have to be discrete because it's not like the median or the mode, where those numbers are discrete. Instead, it's the typical value across the entire distribution. The mean is the most typical value in a set of values, regardless if the values are discrete or continuous. The probabilities of successes in a binomial distribution are discrete, whereas the number of successes measured is continuous.
